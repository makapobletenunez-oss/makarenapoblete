```python
#Problema 1 lista desde -30 a 50

lista_numeros = range(-30,51)
print(lista_numeros)

for x in range(-30,51):
  if(x < 0):
    print(x, "Muy frío")
  elif(x >= 0 and x <= 24):
    print(x, "Templado")
  else:
    print(x, "Caluroso")
```

```python
#Problema 2 Números especiales del robot

lista_numeros = range(1,101)
print(lista_numeros)

for numero in range(1, 101):
    if numero % 10 == 0:
        print(numero, "Bip Bip")
    elif numero % 5 == 0:
        print(numero, "Bip")
```

```python
#Problema 3 Detector de puntaje gamer

puntaje1 = int(input("Ingresa el primer puntaje: "))
puntaje2 = int(input("Ingresa el segundo puntaje: "))
suma_puntajes = puntaje1 + puntaje2

if suma_puntajes < 50:
    print("Resultado: Nivel principiante")
elif 50 <= suma_puntajes < 80:
    print("Resultado: Nivel intermedio")
else:
    print("Resultado: Nivel pro")
```

```python
#Problema 4 Calculador de perfil musical

edad = int(input("Ingresa tu edad: "))
le_gusta_musica = input("¿Te gusta la música? (sí/no): ")
le_gusta_bailar = input("¿Te gusta bailar? (sí/no): ")

le_gusta_musica = le_gusta_musica == "sí"
le_gusta_bailar = le_gusta_bailar == "sí"

if edad >= 18 and le_gusta_musica and le_gusta_bailar:
    print("Eres una estrella de la fiesta")
elif le_gusta_musica and not le_gusta_bailar:
    print("Te gusta la música, pero no tanto bailar")
else:
    print("Tienes tu propio estilo")
```
